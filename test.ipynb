```python
import numpy as np
import matplotlib.pyplot as plt

# Given parameters
period = 288
amplitude = 10
bias = 20
noise_std = 1

# Create time array (let's plot 2 complete periods)
t = np.linspace(0, 2*period, 1000)

# Generate the sinusoidal wave with noise
np.random.seed(42)  # For reproducible results
temp = bias + amplitude * np.sin(2 * np.pi * t / period) + np.random.normal(0, noise_std, len(t))

# Create the plot
plt.figure(figsize=(12, 6))
plt.plot(t, temp, 'b-', linewidth=1, alpha=0.7, label='Temperature with noise')

# Also plot the clean sine wave for comparison
temp_clean = bias + amplitude * np.sin(2 * np.pi * t / period)
plt.plot(t, temp_clean, 'r--', linewidth=2, alpha=0.8, label='Clean sinusoidal wave')

# Add labels and formatting
plt.xlabel('Time (t)', fontsize=12)
plt.ylabel('Temperature', fontsize=12)
plt.title('Sinusoidal Temperature Variation\n(temp = bias + ampl·sin(2πt/period) + noise)', fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend()

# Add parameter info as text
plt.text(0.02, 0.98, f'Period: {period}\nAmplitude: {amplitude}\nBias: {bias}\nNoise std: {noise_std}', 
         transform=plt.gca().transAxes, verticalalignment='top', 
         bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))

plt.tight_layout()
plt.show()
```
